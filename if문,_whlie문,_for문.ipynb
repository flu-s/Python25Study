```python
# while문 강제로 빠져나가기

coffee = 10
money = 300
while money:
    print("돈을 받았으니 커피를 줍니다.")
    coffee = coffee -1
    print("남은 커피의 양은 %d개입니다." % coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지합니다.")
        break
# money가 300으로 고정 whlie money :에서 조건문인 mpney는 0이 아니기 때문에 항상 참이기 때문에 무한 반복되는 무한 루프를 돌게 됨
# whlie문의 내용을 한 번 수행할 때마다 coffee = coffee -1에 의해 coffee의 개수가 1개씩 줄ㄹ어든다
# coffee가 0이 되면 if coffee == 0: 문장에서 coffee == 0이 참이 되므로 if 문 다음 문장 "커피가 다 떨어졌습니다. 판매를 중지합니다."가 출력되고 break 문이 호출되어 while 문을 빠져나가게 된다
```

```python
# for문의 응용

# 총 5명의 학생이 시험을 보았는데 시험 점수가 60점 이상이면 합격이고 그렇지 않으면 불합격이다. 합격인지, 불합격인지 결과를 보여주시오

marks = [90, 25, 67, 45, 80]

number = 0 # 학생에게 붙여 줄 번호
for mark in marks :  # 90, 25, 67, 45, 80을 순서대로 mark에 대입
  number = number +1
  if mark >= 60:
      print("%d번 학생은 합격입니다." %number)
  else:
      print("%d번 학생은 불합격입니다." %number)
```

```python
# for문과 continue문

# for문 안의 문장을 수행하는 도중 continue문을 만나면 for문의 처음으로 돌아간다

# 앞에서 for문 응용 예제를 그대로 사용해서 60점 이상인 사람에게는 축하메세지를 보내고 나머지 사람에게는 아무런 메세지도 전하지 않는 프로그램

marks = [90, 25, 67, 45, 80]

number = 0
for mark in marks:
    number = number +1
    if mark < 60:
        continue
    print("%d번 학생 축하합니다. 합격입니다. " % number)
```

```python
# 리스트 컴프리헨션 사용하기 - list comprehension
# 좀 더 편리하고 직관적인 프로그램을 만들 수 있다

a = [1,2,3,4]
result = []
for num in a:
    result.append(num*3)
print(result)

# 위 예제는 a 리스트의 각 항목에 3을 곱한 결과를 result리스트에 담았다
# 리스트 컴프리헨션을 사용하면 더 간단하다

a = [1,2,3,4]
result = [num *3 for num in a]
print(result)

# 만역 [1,2,3,4]중에서 짝수에만 3을 곱하고 담고 싶다면 리스트 컴프리헨션 안에 'if 조건문' 사용

a = [1,2,3,4]
result = [num *3 for num in a if num % 2 == 0]
print(result)

# 리스트 컴프리헨션의 기본 문법
# 'if 조건문' 부분은 생략 가능

# [표현식 for 항목 in 반복_가능_객체 if 조건문]

# for문을 2개 이상 사용하는 것도 가능

# [표현식 for 항목1 in 반복_가능_객체1 if 조건문1
#        for 항목2 in 반복_가능_객체2 if 조건문2
#        for 항목n in 반복_가능_객체n if 조건문n]

# 구구단의 모든 결과를 리스트레 담고 싶으면 리스트 컴프리헨션 사용해 간단하게 구현가능

result = [x*y for x in range(2,10)
              for y in range(1,10)]
print(result)
```

```python
# enumerate 함수 사용하기

# 리스트의 순서(인덱스)와 값을 함께 구하고 싶을 때는 enumerate 함수를 사용하면 편하다

fruits = ['apple', 'banana', 'ornage']
for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")

# enumerate는 0부터 시작하는 인덱스 번호를 자동으로 생성해준다
# 시작 번호를 변경하고 싶으면 예제와 같이 사용할 수 있다

fruits = ['apple', 'banana', 'orange']
for i, fruit in enumerate(fruits, 1): # 1부터 시작
    print(f"{i}:{fruit}")
```

```python
from os import name
# zip 함수로 여러 리스트 함께 순회하기

# 두개 이상의 리스트를 동시에 순회하고 싶을 때 zip함수 사용

names = ['홍길동','김철수','이영희']
scores = [85, 92, 78]
for name, score in zip(names, scores):
    print(f"{name} : {score}점")

# 3개 이상의 리스트도 함께 순회할 수 있다

names = ['홍길동', '김철수', '이영희']
korean = [85, 92, 78]
english = [90, 88, 95]
for name, kor, eng in zip(names, korean, english):
    print(f"{name}: 국어 {kor}점, 영어 {eng}점")
```
